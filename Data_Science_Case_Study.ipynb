# 2. Data Pre-Processing: 

```python
#Excel dosyasini yukleme islemi
df = pd.read_excel('side_effect_data.xlsx')
```

```python
#Eksik degerleri sayma islemi
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
#Bosluklari NaN a donusturme islemi
import numpy as np
df.replace(' ', np.nan, inplace=True)
```

```python
#SimpleImputer ile Eksik verileri tamamlama islemi
from sklearn.impute import SimpleImputer

sayisal_sutunlar= ['Kilo', 'Boy'] 
imputer = SimpleImputer(strategy='median')  
df[sayisal_sutunlar] = imputer.fit_transform(df[sayisal_sutunlar])
```

```python
kategorik_sutunlar = ['Kiz Kardes Kronik Hastaliklari','Erkek Kardes Kronik Hastaliklari',"Cinsiyet", "Il", "Alerjilerim",'Kan Grubu','Kronik Hastaliklarim', 'Baba Kronik Hastaliklari',
                    'Anne Kronik Hastaliklari']
imputer = SimpleImputer(strategy='most_frequent')
df[kategorik_sutunlar] = imputer.fit_transform(df[kategorik_sutunlar])
```

```python
#Dosyayi yeni excel e kaydetme islemi
df.to_excel('yeni_data.xlsx', index=False)
yeni_df = pd.read_excel('yeni_data.xlsx')
```
